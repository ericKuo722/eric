Let $X\,{\sim}\,\mathcal{IG}\left(\theta: \left(\mu, \lambda\right)\right),\;\,\forall\, \mu>0 \text{ and } \lambda>0$. This means
the random varible $X$ follows the **inverse Gaussian distribution** with the set $\left(\theta: \left(\mu, \lambda\right)\right)$ acting as the parameters of said distribution. Given that we observe a sample of size $n$ that is independently and identically distributed from this distribution ([i.i.d](https://en.wikipedia.org/wiki/Independent_and_identically_distributed_random_variables)), $\mathbf{x}=$ $\left(x_{1}, \ldots, x_{n}\right)$, please find the [maximum likelihood estimate](https://en.wikipedia.org/wiki/Maximum_likelihood_estimation) for $\mu$ and $\lambda$, that is $\mu_{\text{MLE}}$ and $\lambda_{\text{MLE}}$. The probability density function (**PDF**) is as follows:

$$
f(x \mid \mu, \lambda)=\left\{\begin{array}{cc}
\left(\frac{\lambda}{2 \pi x^{3}}\right)^{1 / 2} e^{\frac{-\lambda(x-\mu)^{2}}{2 \mu^{2} x}}, & x>0 \\
0, & x \leq 0
\end{array}\right.
$$

### ANSWER

**1. Likelihood** 

$$
p(x \mid \mu ,\lambda)=\prod_{i=1}^{n}p(x_{i} \mid \mu ,\lambda)
$$



$$
p(x \mid \mu ,\lambda) = \left(\frac{\lambda}{2 \pi (x_{1})^{3}}\right)^{1 / 2} e^{\frac{-\lambda(x_{1}-\mu)^{2}}{2 \mu^{2}x_{1}}} \cdot  \left(\frac{\lambda}{2 \pi (x_{2})^{3}}\right)^{1 / 2} e^{\frac{-\lambda(x_{2}-\mu)^{2}}{2 \mu^{2}x_{2}}}\cdot\cdot\cdot \left(\frac{\lambda}{2 \pi (x_{n})^{3}}\right)^{1 / 2} e^{\frac{-\lambda(x_{n}-\mu)^{2}}{2 \mu^{2} x_{n}}}
$$

**2. Negative log-likelihood**

$$
\frac{-n}{2}\log{ \frac{\lambda }{2\pi x^{3} } }+\sum_{i=1}^{n} \left ( \frac{\lambda\left (  x-\mu \right )^2  }{2\mu^2x }  \right ) 
$$

**3. MLE**

$$
\frac{d \ell(x \mid \mu ,\lambda)}{d \mu } = \lambda \sum_{i=1}^{n}  \frac{(x-\mu)}{\mu^3} 
$$
$$
\frac{d \ell(x \mid \mu ,\lambda)}{d \lambda } = \frac{(-n)}{2x} + \sum_{i=1}^{n}\frac{(x-\mu)^2}{2\mu^2x}   
$$

````r
Plot.CLT <-function(n, dist){
    
    output_all <- NULL
    
    for (i in seq(1,1000)){
        random <- sample(1:5, n, replace = T, prob = c(0.35, 0.05, 0.15, 0.05, 0.4))
        
        if (dist == 'sum'){
            output <- round(sum(random),5)
        }
        else{
            output <- round(mean(random),5)
        }
        output_all <- append(output_all, output)
    }
    
    return(output_all)
}
````

````r
for(i in c(5, 30, 100)){
    
    Summation <- Plot.CLT(i, 'sum')
    Mean <- Plot.CLT(i, 'mean')
    
    hist(Summation, xlab  = 'Summation', ylab = 'Density', 
         main = paste('Summation Distribution n:', i), freq = F)
    lines(density(Summation), col = 'red', lwd = 3)
    
    hist(Mean, xlab  = 'Mean', ylab = 'Density', main = paste('Mean Distribution n:', i), freq = F)
    lines(density(Mean), col = 'red', lwd = 3)
}
````

````r
# ANSWER BLOCK

# Least squared estimator function
lsq <- function(dataset, predictor){
    # INSERT YOUR ANSWER IN THIS BLOCK
    
    SSXX <- mean(dataset[, predictor]^2) - mean(dataset[, predictor])^2 
    SSXY <- mean(dataset[,predictor]*dataset[,'consc_lev'])-mean(dataset[,predictor])*mean(dataset[,'consc_lev'])
    # Get the final estimators
    beta_1 <- SSXY/SSXX
    beta_0 <- mean(dataset[,'consc_lev']) - beta_1*mean(dataset[, predictor]) 
    
    # Return the results:
    return(paste0('E[consc_lev]=', beta_0,'+', beta_1,'*',predictor))
}
print(lsq(train, 'input'))
print(lsq(train, 'v_pyr'))
````
